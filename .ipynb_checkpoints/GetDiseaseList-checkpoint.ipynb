```python
#coding:utf-8
import requests
import json
import string
import time
import bs4
from bs4 import BeautifulSoup
import re
def getHTMLText(url):
    try:
        #print("1")
        header={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.80 Safari/537.36'}
        r = requests.get(url,timeout = 30, headers = header)
        #print("2")

        return r.text
    except:
        return "失败了"

    
def getUrls(html):
    
    soup = BeautifulSoup(html, "lxml")
    tex = soup.find(attrs={'class':'bottom-list'})
    #print(tex)
    pattern1 = re.compile(r"<a.+?href=\"(.+?)\".*>")
    pattern2 = re.compile(r"</i>(.+?)</a>")
    
    Disease_Dict = {}
    list_type = tex.findAll('a')
    #print(list_type)
    for disease in list_type:
        link_list = re.findall(pattern1, str(disease))
        link = "".join(link_list)
        name_list = re.findall(pattern2, str(disease))
        name = "".join(name_list)
        Disease_Dict[name] = link
    return Disease_Dict
    





def main():
    url = "https://m.ew86.com/jibing/jibing/pinyin-"
    Disease_Dict = {}
    for i in string.ascii_uppercase:
        try:
            html = getHTMLText(url+i)
            #print(html)
            print("\n\n\n")
            #-----------获取疾病集合-----------#
            disease = getUrls(html)
            
            #print(Disease_Dict)
        except:
            pass
        finally:
            Disease_Dict.update(disease)
            print(Disease_Dict)
    
    jsObj = json.dumps(Disease_Dict, ensure_ascii=False)
    fileObject = open('./data/LinkAndDisease.json', 'a+', encoding='utf8')
    fileObject.write(jsObj)
    fileObject.close()
    print("结束")
    
main()
```
